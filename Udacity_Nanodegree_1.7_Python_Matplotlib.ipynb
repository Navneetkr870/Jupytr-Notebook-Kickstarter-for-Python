## 1. Introduction

Generally visualisations within Python have a trade-off between **Flexibility** and **Productivity**.

+ **Matplotlib** : a versatile library for visualizations, but it can take some code effort to put together common visualizations.

+ **Seaborn** : built on top of matplotlib, adds a number of functions to make common statistical visualizations easier to generate.

+ **pandas** : while this library includes some convenient methods for visualizing data that hook into matplotlib, we'll mainly be using it for its main purpose as a general tool for working with data.

## 2. Bar Charts

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
pokemon = pd.read_csv('./data/Matplotlib/pokemon.csv')
df=pokemon.copy()
print(pokemon.shape)
pokemon.head(5)
```

Showing how many species were introduced in every generation

```python
sns.countplot(data=pokemon, x='generation_id');
```

Given the colors are coming different but our plot doesn't really require to be in different colors. Hence we would fix it to one color

```python
base_color = sns.color_palette()[0]
base_color
```

```python
sns.countplot(data=pokemon, x='generation_id', color = base_color);
```

Next on, we want to order this bar chart as per the counts :

1. We first find the indexes of the ordering

```python
gen_order = pokemon['generation_id'].value_counts().index
gen_order
```

2. We pass this on to the plot

```python
sns.countplot(data=pokemon, x='generation_id', color = base_color, order=gen_order);
```

Let's try it on another grouping

```python
gen_order = pokemon['type_1'].value_counts().index
gen_order
sns.countplot(data=pokemon, x='type_1', color = base_color, order=gen_order)
plt.xticks(rotation=90);
```

```python
sns.countplot(data=pokemon, y='type_1', color = base_color, order=gen_order);
```

```python
# create the plot
base_color = sns.color_palette()[0]
sns.countplot(data = pokemon, x = 'type_1', color = base_color)

# add annotations
n_points = pokemon.shape[0]
cat_counts = pokemon['type_1'].value_counts()
locs, labels = plt.xticks() # get the current tick locations and labels

# loop through each pair of locations and labels
for loc, label in zip(locs, labels):
#     print('Location : {}, Label: {}'.format(loc,label)) Used for debugging
    
    # get the text property for the label to get the correct count
    count = cat_counts[label.get_text()]
    pct_string = '{:0.1f}%'.format(100*count/n_points)

    # print the annotation just below the top of the bar
    plt.text(loc, count+2, pct_string, ha = 'center', color = 'b', fontsize=8 )
    
    plt.xticks(rotation=90)
```

### 1. Scatter Plot

```python
# Scatterplot arguments
sns.lmplot(x='attack', y='defense', data=pokemon,
           fit_reg=False,  # No regression line 
           hue='type_1')   # Color by evolution stage 
```

Scatterplot is not available in seaborn but we've used linear regression to plot this. Also while seaborn will give us a template, you will have to use Matplotlib to customize the charts

### 2. Customizing Seaborn using Matplotlib

Seaborn is a high-level interface to Matplotlib. From our experience, Seaborn will get you most of the way there, but you'll sometimes need to bring in Matplotlib.

1. First, invoke your Seaborn plotting function as normal.
2. Then, invoke Matplotlib's customization functions. 

In this case, we'll use its ylim() and xlim() functions to limit the axes of the charts

```python
# Plot using Seaborn
sns.lmplot(x='attack', y='defense', data=pokemon,
           fit_reg=False,  # No regression line 
           hue='type_1')   # Color by evolution stage 

# Tweak using Matplotlib
plt.ylim(-100, 300)
plt.xlim(-100, 300)

plt.grid(b=True)
plt.title("Pokemon's Attack vs. Defense")
```

### 3. Role of Pandas

As this is a Visualisation resource but we would look into how Pandas Dataframe can help us bring better visualisations in place

```python
# Boxplot
sns.boxplot(data=pokemon)
plt.xticks(rotation=90)
```

While let's say we have to only include Pokeman related stats then we can remove columns like - id, generation_id

We can do that using Dataframes rather than doing directly within seaborn

```python
# Pre-format DataFrame
stats_df = pokemon.drop(['id', 'generation_id', 'weight'], axis=1)
 
# New boxplot using stats_df
sns.boxplot(data=stats_df)
plt.xticks(rotation=90)
```
